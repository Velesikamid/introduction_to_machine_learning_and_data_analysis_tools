```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # эта команда устанавливает стиль диаграммы seaborn по умолчанию
```

```python
# Подключение к файлу с данными
df_used_cars = pd.read_csv("https://raw.githubusercontent.com/trisha00001/EDA/main/bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9, 6)) #размер диаграммы
# создать столбчатую диаграмму'bar'.
plt.bar(x = df_used_cars["Brand"], #подпишем ось x
        height = df_used_cars["Cars Listings"], #подпишем ось y
        color = "midnightblue") # определим цвет
plt.xticks(rotation = 45, fontsize = 13) # повернуть и отформатировать надписи по оси x
plt.yticks(fontsize = 13) # формат шрифта y
plt.title("Объявления об автомобилях​ в зависимости от бренда", fontsize = 16, fontweight = "bold") #Формат заголовка
plt.ylabel("Количество просмотров", fontsize = 13 ) #заголовок для оси y
plt.savefig("Cars_Bar.png") # Сохраним файл
plt.show() # в зависимости от того, какую среду вы используете, без этой строки ваша диаграмма может отображаться неправильно.
```

```python
plt.figure(figsize = (9, 6))
plt.barh(y = df_used_cars["Brand"], #подпись y
        width = df_used_cars["Cars Listings"], #подпись x
        color = "midnightblue")
plt.xticks( fontsize = 13)
plt.yticks( fontsize = 13)
plt.title("Объявления об автомобилях​ в зависимости от бренда", fontsize = 16, fontweight = "bold")
plt.xlabel("Количество просмотров", fontsize = 13 )
plt.tight_layout()#tight_layout автоматически настраивает параметры подграфика таким образом, чтобы подграфик(и) вписывался в область рисунка
plt.savefig("Cars_Bar.png")
plt.show()
```

```python
# load data. Number of cars based on their engine fuel types. Four engine fuel types in total : Diesel, Gas, Petrol and Other.
df_fuel_engine_type = pd.read_csv("https://raw.githubusercontent.com/trisha00001/EDA/main/pie_chart_data.csv")
```

```python
# Установите цветовой цикл matplotlib, используя палитру seaborn
sns.set_palette('colorblind')
```

```python

# Метки добавляются по мере увеличения количества автомобилей в каждой категории в виде процентов.
# Дополнительное форматирование меток с помощью textprops. Обратите внимание, что это словарь.

plt.figure(figsize = (10, 8))
plt.pie(df_fuel_engine_type['Number of Cars'],
        labels = df_fuel_engine_type['Engine Fuel Type'].values,
        autopct = '%.2f%%',
        textprops = {'size' : 'x-large',
                     'fontweight' : 'bold',
                     'rotation' : 'horizontal',
                     'color' : 'w'})
plt.legend()
plt.title('Срез машин в зависимости от потребления топлива', fontsize = 18, fontweight = 'bold')
plt.show()
```

```python
df_fuel_engine_types = pd.read_csv("https://raw.githubusercontent.com/trisha00001/EDA/main/stacked_area_chart_data.csv")
df_fuel_engine_types
```

```python
#  Создайте диаграмму с наложенными областями с помощью pyplots 'stackplot'. На оси x у нас есть временная линия - с 1982 по 2016 год.
# На оси y у нас есть три категории: "Дизельное топливо", "Бензин" и "Газ", расположенные друг над другом.
# Обратите внимание, что этот порядок выбран специально для того, чтобы категории были от самых больших к самым маленьким.
# Это помогает нам сравнить размер категорий, поскольку люди испытывают трудности с определением размера непрямоугольных фигур.

# Предоставьте список цветов, чтобы каждая категория могла иметь определенный цвет.
# Названия цветов должны отображаться в том же порядке, в каком отображаются категории наложенных областей.
colors = ["#011638", "#7e2987", "#ef2026"]
# список меток для легенды. Названия должны отображаться в том же порядке, в каком отображаются категории участков с наложенной областью.
labels = ["Дизель", "Бензин", "Газ"]
sns.set_style("white") # Используйте "белую" тему seaborn, чтобы ввести белый фон вместо серого по умолчанию.
plt.figure(figsize = (12, 6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor = 'none')# метод, показатель модуля библиотек matplotlib библиотека используется для получения кромки цвет фигуры прямоугольник.
plt.xticks(df_fuel_engine_types["Year"], rotation = 45) # Включите метки по оси x для каждого года и поверните метки на 45 градусов.
plt.legend(labels = labels, loc = "upper left") # Добавьте условное обозначение и укажите его местоположение на диаграмме.
plt.ylabel("Количество машин", fontsize = 13)
plt.title("Соотношение видов моторного топлива (1982 - 2016)", fontsize = 14, weight = "bold")
sns.despine() # Удалите верхнюю и правую границы диаграммы.
plt.show()
```

```python
df_spx_ftse_00_10 = pd.read_csv("https://raw.githubusercontent.com/trisha00001/EDA/main/line_chart_data.csv")
df_spx_ftse_00_10
```

```python

# Преобразовать "Дату" в формат datetime. В противном случае это будет рассматриваться как простая строка
# поэтому операции или преобразования в столбце даты не приведут к желаемому результату.
# Это важнейший шаг в любом анализе временных рядов.
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
```

```python
# regplot принимает x и y в различных форматах. Для regplot данные должны быть строками в так называемой «длинной форме»
#plt.figure(figsize = (10, 8))
sns.set(rc = {'figure.figsize': (9,6)}) # фиксирует размер с помощью словаря
sns.regplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,
            scatter_kws = {'color': 'k'}, # цвет точек
            line_kws = {'color': 'red'}) # цвет линии регрессии
plt.xlabel("Расходы на рекламу в (000 $)")
plt.ylabel("Продажм (000 шт.)")
plt.title("Влияние расходов на рекламу на продажи", fontsize = 14, weight = "bold")
plt.show()
```

```python
from matplotlib.ticker import PercentFormatter #преобразует значения в процентный формат
```

```python
df_kdnuggets = pd.read_csv("https://raw.githubusercontent.com/trisha00001/EDA/main/bar_line_chart_data.csv") #read in the KDnuggets survey data file
```

```python
df_kdnuggets #датафрейм, который мы будем использовать для создания комбинированного графика. Две диаграммы будут иметь общую ось X, которая соответствует году.
#На гистограмме количество участников будет указано на основной оси X (в левой части диаграммы).
# На линейной диаграмме пользователи Python будут указаны в процентах на вторичной оси Y (в правой части диаграммы).
```

```python
# Создание комбинированной диаграммы. Здесь мы используем фигуру с осями.
# Первая ось или подграфик представляет собой гистограмму, вторая — линейную диаграмму.
fig, ax = plt.subplots(figsize = (10, 7))

#создание и стилизация гистограммы
ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],
       color = "k")
ax.set_ylabel("Количество пользователей",
              weight = "bold")
ax.tick_params(axis = "y",
               width = 2,
               labelsize = "large")
ax1 = ax.twinx()
#изменение вторичной оси Y для отображения процентов по шкале от 0% до 100%
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
#создание и стилизация линейного графика
ax1.plot(df_kdnuggets["Year"],
         df_kdnuggets["Python Users"],
         color = "#b60000",
         marker = "D")
ax1.set_ylabel("Пользователи Python",
               color = "#b60000",
               weight = "bold")
ax1.tick_params(axis = "y",
                colors = "#b60000",
                width = 2,
                labelsize = "large")
ax.set_title("Опрос пользователей Python (2012 - 2019)", fontsize = "14", weight = "bold")
plt.show()
```
